```python
from pprint import pformat
from csv import DictReader

def read_labels(filename):
    labels_dim_1 = {}
    labels_dim_2 = {}
    with open(filename, 'rt') as f:
        reader = DictReader(f, delimiter = '\t')
        
        for line in reader:
            docname = line['Document']
            dim1 = line['Dimension 1']
            dim2 = line['Dimension 2']
            
            labels_dim_1[docname] = dim1
            labels_dim_2[docname] = dim2
    return labels_dim_1, labels_dim_2


    
    
labels_dim_1, labels_dim_2 = read_labels(labels_file_path)
labels_dim_1
```

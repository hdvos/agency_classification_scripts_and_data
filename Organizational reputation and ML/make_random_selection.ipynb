III. Drawing a random sample 

1. Please draw a random sample of 400 documents of all the agencies in our sample (34) for all the years excluding 2019 (because it has been coded already), excluding BBI, ECSEL, Cleansky. 

```python
all_agencies.pop(all_agencies.index("BBI"))
all_agencies.pop(all_agencies.index("CLEANSKY"))
all_agencies.pop(all_agencies.index("ECSEL"))
all_agencies
# all_agencies.pop(all_agencies.index("BBI"))
```

```python
def make_agency_selector(row):
    agency=row.agency
    return agency in all_agencies
```

```python
selector = data.apply(make_agency_selector, axis=1)
data = data[selector]
```

```python
data.head()
```

```python
data = data[data.year != 2019]
```

```python
data.head()
```

```python
selections = []
```

```python
from math import ceil
total_docs = data.shape[0]
sample = 400
print(total_docs)
for agency in all_agencies:
    agency_data = data[data.agency==agency]
    n_docs = agency_data.shape[0]
    percentage_of_total = n_docs/total_docs
    if n_docs == 0:
        continue
    selection_size = ceil(max(1, percentage_of_total*sample))
    
    selection = agency_data.sample(selection_size)
    selections.append(selection)
    
    print(agency, selection_size)
```

```python
total_selection = pd.concat(selections)
total_selection
```

```python
total_selection.columns
```

```python
total_selection = total_selection[['id', 'date', 'title', 'text']]
total_selection
```
